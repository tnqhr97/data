```python
import pandas as pd
import numpy as np
from pandas import datetime
from matplotlib import pyplot as plt

"""
Load AirQualityUCI Data
"""

def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

input_file = 'C:/data/AirQualityUCI_refined.csv'

df = pd.read_csv(input_file,
                 index_col=[0],
                 parse_dates=[0],
                 date_parser=parser)

df.head()
```

```python
pt = df["PT08.S3(NOx)"].copy()
print(pt.head())
pt.plot(figsize=(10,5))
```

```python
pt.interpolate(inplace=True)

plt.figure(figsize=(10,5))
plt.plot(df["PT08.S3(NOx)"], zorder=2)
plt.plot(pt, zorder=1)
```

```python
plt.boxplot(pt)
plt.xlabel("PT08.S3(NOx)")
```

```python
"""
IQR-based Outlier Detection
"""

# Q1, Q2(median), Q3
q1 = pt.quantile(0.25)
median = pt.quantile(0.5)
q3 = pt.quantile(0.75)
print(q1,median,q3)
```

```python
# IQR, upper_fence, lower_fence
iqr = q3 - q1 
upper_fence = q3 + 1.5*iqr
lower_fence = q1 - 1.5*iqr
print(iqr,upper_fence, lower_fence)
```

```python
# Filtering the outliers
outliers = pt.loc[(pt > upper_fence)|(pt < 0)]
outliers
```

```python
# Mask for outliers
# True -> 이상치
mask = pt.index.isin(outliers.index)
mask[:50]
```

```python
# Removing the outliers
pt_refined = pt.copy()
pt_refined[mask] = np.nan
print(pt_refined[mask])
```

```python
# Linear interpolation for reconstructing outliers removed.
pt_refined.interpolate(inplace=True)  
pt_refined.plot()
```

```python
"""
Detecting Outliers with Z-Scores
"""
import seaborn as sns
sns.distplot(pt)
```

```python
# Mean, Standard deviation
mean = np.mean(pt)
std = np.std(pt)
print(mean,std)
```

```python
# Calculate Z-scores for each data points
outliers = []
thres = 3   # z스코어의 임계치

for i in pt:
    z_score = (i-mean)/std
    if (np.abs(z_score) > thres):
        print(z_score)
        outliers.append(i)
```

```python
# Simplified version of filtering outliers
outliers = pt.loc[np.abs((pt-mean)/std) > 3]
outliers
```

```python
# Mask for outliers
mask = pt.index.isin(outliers.index)
mask[:50]
```

```python
# Comparison of distributions before/after outlier removal
sns.distplot(pt, axlabel='PT08.S3(NOx)', label='original')
sns.distplot(pt[~mask], label='outliers removed')
plt.legend(loc='best')

# [exer] Adjust thres
```
